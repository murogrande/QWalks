## Evolution on path graph for fixed time, global vs local interaction

Basic parameters. Note that *dim* should be an odd number. Otherwise the middle point *midpoint* is not unique. Matrix
In the case of the PathGraph:*adjacency* can be alternatively generated as *adjacency = spdiagm((ones(dim-1), ones(dim-1)), (-1, 1))*.
Using another graph like WheelGraph, the adjacency matrix is not the same as the PathGraph adjacency matrix.
In general, one has to use the *adjacency_matrix*

```julia
dim = 15 # odd for unique middle point. Also, we can use small number of nodes for ploting issues
midpoint = ceil(Int, dim/2)
w = 0.5 # we are using the QSW equation with parameter w=0.5. 
timepoint = 10
grafo = PathGraph(dim)
#grafo = WheelGraph(dim)
adjacency = adjacency_matrix(grafo);
```

```julia
gplot(grafo, nodelabel=vertices(grafo))
```

```julia
# Adjacency matrix of the PathGraph
adjacency
```

Quantum stochastic walk: global interaction case. Note that *evolve_generator* requires a list of Lindblad operators, hence we need extra brackets around the second argument.

```julia
op_global = evolve_generator(adjacency, [adjacency], w)
rho_global = evolve(op_global, proj(midpoint, dim), timepoint);
#rho_global = evolve(op_global, proj(1, dim), timepoint);
```

Quantum stochastic walk: local interaction case. Note that here *lind_local* is already a list of Lindbladian operators, hence we do not need extra brackets, contrary to the global interaction example.

```julia
lind_local = local_lind(adjacency)
op_local = evolve_generator(adjacency, lind_local, w)
rho_local = evolve(op_local, proj(midpoint, dim), timepoint);
#rho_local = evolve(op_local, proj(1, dim), timepoint);
```

To plot the result of the cannonical measurement, we take a diagonal of the states. Note that both *rhoglobal* and *rholocal* are complex matrices, hence we need to take the real part only. Argument *positions* is generated assuming that *midpoint* (the initial position) is at 0. Note that we have very heavy tails in the global interaction case, which confirms fast propagation in this model.

```julia
# Cell for testing purposes 
#adjacencyd
```

```julia
#positions = (collect(1:dim) .- midpoint)
positions = (collect(1:dim))
plot(positions, real.(diag(rho_local)), "k",label="local")
plot(positions, real.(diag(rho_global)), "b",label="global")
xlabel("position")
ylabel("probability")
PyPlot.grid()
legend()
show()
```

## Evolution on path graph for fixed time, propagation analysis

Basic parameters. Note that *dim* should be an odd number. Otherwise the middle point *midpoint* is not unique. In the case of a PathGraph, Matrix *adjacency* can be alternatively generated as *adjacency = spdiagm((ones(dim-1), ones(dim-1)), (-1, 1))*.
In general, it is important to use *adjacency_matrix*

```julia
dim = 15 #odd for unique middle point
midpoint = ceil(Int, dim/2)
w = 0.5
#timepoints = collect(0:2.5:50)
timepoints = collect(0:0.5:10)
adjacency = adjacency_matrix(PathGraph(dim));
```

Preparation of the global and the local interaction operators.

```julia
op_global = evolve_generator(adjacency, [adjacency], w)
lind_local = local_lind(adjacency)
op_local = evolve_generator(adjacency, lind_local, w);
```

Calculating resulting states. Note that comparing to global operators, density states usually are small matrices. Hence we can store those. Furthermore, note that here we use variant of *evolve* function, whcih enables us to compute the same evolutions for different time points.

```julia
global_states = evolve(op_global, proj(midpoint, dim), timepoints)
local_states = evolve(op_local, proj(midpoint, dim), timepoints);
```

Calculation of the second moment.

The second moment in quantum walks refers to the variance of the position distribution of a quantum particle during a quantum walk on a graph. It is a measure of how spread out the position distribution is and provides insight into the localization and delocalization properties of the quantum walk. The second moment is usually calculated by taking the expected value of the square of the position operator, with respect to the state of the quantum particle at a given time. In quantum walks on graphs, the second moment can be used to study the behavior of quantum algorithms and to design new quantum algorithms that exploit the properties of quantum walks.

```julia

#diag(rho_global)
```

```julia
secmoment_global = Float64[]
secmoment_local = Float64[]
#positions = (collect(1:dim) .- midpoint)
positions = (collect(1:dim))

for (rho_global, rho_local) = zip(global_states, local_states)
  push!(secmoment_global, sum(positions.^2 .* diag(rho_global)))
  push!(secmoment_local, sum(positions.^2 .* diag(rho_local)))

end
```

Plots of the results. Note that second moment for global interaction case grows ballistically.

When the second moment of a quantum walk grows ballistically, it means that the variance of the position distribution of the quantum particle increases linearly with time, as if the particle were moving ballistically. This growth pattern is a hallmark of delocalization, where the quantum particle spreads out rapidly over the graph and explores many different vertices. Ballistic growth of the second moment indicates that the quantum walk is spreading out rapidly and efficiently over the graph, which can be useful for tasks such as searching and exploring large networks. However, the ballistic growth can also result in exponential growth of the standard deviation, which can be a drawback in certain applications where it is important to keep the position distribution of the quantum particle tightly localized.

```julia
plot(timepoints, secmoment_local, "k",label="local")
plot(timepoints, secmoment_global, "b",label="global")
xlabel("t")
ylabel(L"\mu_2")
legend()
PyPlot.grid()
show()
```

If the second moment of a quantum walk is constant, it means that the variance of the position distribution of the quantum particle remains the same over time. This indicates that the quantum particle is not spreading out over the graph and is instead localized in a specific region. A constant second moment is a signature of localization, where the quantum particle remains confined to a small region of the graph and does not explore other parts of the graph. In quantum walks on graphs, localization can be caused by various factors, such as interference between different paths, symmetry in the graph, or certain structures in the graph. Constant second moment implies that the quantum walk is not spreading out, which can be useful for tasks where it is important to keep the quantum particle localized, such as in quantum algorithms that require precision.


If the second moment of a quantum walk oscillates, it means that the variance of the position distribution of the quantum particle alternates between increasing and decreasing over time. This behavior indicates that the quantum particle is delocalizing and localizing repeatedly, and the oscillations in the second moment reflect these localization-delocalization transitions. In quantum walks on graphs, oscillations in the second moment can be caused by various factors, such as interference between different paths, symmetries in the graph, or certain structures in the graph. The oscillations can also be used to study the behavior of quantum algorithms and to design new quantum algorithms that exploit the properties of quantum walks. Oscillations in the second moment can have both advantages and disadvantages, depending on the specific application and the desired behavior of the quantum walk.


When the second moment of a quantum walk grows initially with a peak and then oscillates, it means that the variance of the position distribution of the quantum particle starts off by increasing rapidly, reaching a maximum value, and then alternating between increasing and decreasing over time. This behavior reflects a combination of delocalization and localization, where the quantum particle spreads out rapidly at the beginning and then becomes confined to a specific region, oscillating between delocalization and localization. The peak in the second moment can indicate that the quantum particle has explored a large part of the graph initially and then become localized. In quantum walks on graphs, this behavior can be caused by various factors, such as interference between different paths, symmetries in the graph, or certain structures in the graph. The combination of rapid growth and subsequent oscillations in the second moment can have both advantages and disadvantages, depending on the specific application and the desired behavior of the quantum walk.
